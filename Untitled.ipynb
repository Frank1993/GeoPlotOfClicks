```python
AdIds = ["10978732666","11439993499","13618219706"]
for AdId in AdIds:
    df = pd.read_csv('D:\Projects\PClicks\Data\WBDataGeoOfAds_%s.txt'%AdId,sep='\t',names = ["AdId",
           "AdTitle",
           "AdTargetRadius",
           "LocationCity",
           "LocationMetroArea", "Latitude", "Longtitude", "Impresions","Clicks","GoodClicks"])

    def getColor(row):
        impresions = row["Impresions"]
        clicks = row["Clicks"]
        goodClicks = row["GoodClicks"]

        if goodClicks == 1:
            color = "blue"
        elif clicks == 1:
            color = "green"
        else:
            color = "red"

        return color

    df["Color"] = df.apply(getColor,axis = 1)

    def getOpacity(row):
        impresions = row["Impresions"]
        clicks = row["Clicks"]
        goodClicks = row["GoodClicks"]

        if goodClicks == 1:
            opacity = 1.0
        elif clicks == 1:
            opacity = 1.0
        else:
            opacity = 0.2

        return opacity

    df["Opacity"] = df.apply(getOpacity,axis = 1)


    #df = df.sample(frac=0.1,axis = 0)


    df['text'] = df['AdId']# + '' + df['Latitude'] + ', ' + df['Longtitude'] 






    #scl = [ [0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
        #[0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"] ]

    data = [ dict(
            type = 'scattergeo',
            locationmode = 'USA-states',
            lon = df['Longtitude'],
            lat = df['Latitude'],
            text = df['text'],
            mode = 'markers',
            marker = dict( 
                size = 7, 
                opacity = df["Opacity"],
                reversescale = True,
                autocolorscale = False,
                symbol = 'circle',
                line = dict(
                    width=1,
                    color='rgba(102, 102, 102)'
                ),
                #colorscale = scl,
                #cmin = 0,
                color = df['Color'],

                #colorbar=dict(
                    #title="Impresion and Clicks"
                #)
            ))]

    layout = dict(
            title = 'Geo Distribution of Ads %s: \n %s'% (AdId,df.loc[0,"AdTitle"]),
            colorbar = True,   
            geo = dict(
                scope='usa',
                projection=dict( type='albers usa' ),
                showland = True,
                landcolor = "rgb(250, 250, 250)",
                subunitcolor = "rgb(38, 9, 12)",
                countrycolor = "rgb(38, 9, 12)",
                countrywidth = 1,
                subunitwidth = 1        
            ),
        )

    fig = dict( data=data, layout=layout )
    py.iplot( fig, validate=False, filename='Geo Distribution of Ads %s'%AdId)
```

```python
import plotly.plotly as py
py.sign_in('Frank1993', 'kgkpACBeHU2lf1yFoWEW')
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_february_us_airport_traffic.csv')
df.head()

df['text'] = df['airport'] + '' + df['city'] + ', ' + df['state'] + '' + 'Arrivals: ' + df['cnt'].astype(str)

scl = [ [0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
    [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"] ]

data = [ dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = df['long'],
        lat = df['lat'],
        text = df['text'],
        mode = 'markers',
        marker = dict( 
            size = 8, 
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'square',
            line = dict(
                width=1,
                color='rgba(102, 102, 102)'
            ),
            colorscale = scl,
            cmin = 0,
            color = df['cnt'],
            cmax = df['cnt'].max(),
            colorbar=dict(
                title="Incoming flightsFebruary 2011"
            )
        ))]

layout = dict(
        title = 'Most trafficked US airports<br>(Hover for airport names)',
        colorbar = True,   
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5        
        ),
    )

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='d3-airports' )
```
